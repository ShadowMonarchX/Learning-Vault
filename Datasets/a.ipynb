```python
# Series from list
s1 = pd.Series([10, 20, 30])
print(s1)

# Series from dictionary
s2 = pd.Series({'a': 1, 'b': 2, 'c': 3})
print(s2)

# DataFrame from dictionary
df1 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
print(df1)

# DataFrame from NumPy array
data = np.array([[1, 2], [3, 4]])
df2 = pd.DataFrame(data, columns=['A', 'B'])
print(df2)


x1 = pd.Series([10,20,30,40,50,60])
print(x1)

print("........")

x2 = pd.Series({'a':2,'b':3})
print(x2)

print("........")


x3 = pd.DataFrame({'Name':['a','b','c'],'age':[1,2,3]})
print(x3)

print("........")

x4 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
x5 = pd.DataFrame(x4, columns=['A','B','C','D','E'])
print(x5)
```

```python
# Download Titanic dataset first or use seaborn's built-in one
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
print(titanic_df.head())
print(titanic_df.tail())
print(titanic_df.shape)
print(titanic_df.info())
print(titanic_df.describe())
```

```python
# Single column
print(titanic_df['Name'])
print("........")
# Multiple columns
print(titanic_df[['Name', 'Age']])
print("........")
# Select rows by index
print(titanic_df.iloc[0])         # First row
print("........")
print(titanic_df.iloc[0:5])       # First 5 rows
print("........")

# Select rows by label
print(titanic_df.loc[0:5, ['Name', 'Sex', 'Survived']])
```
